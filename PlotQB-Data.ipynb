## Notebook for plotting quantum battery data

In this notebook, we provide the scripts to obtain figures 2-4 of our paper: High-performance Kerr quantum battery, published in Appl. Phys. Lett. (2023).

```python
from pylab import cm
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data

# Fig. 2
time, ener0, ener1, ener2, ener3, ener4, ener5 = np.loadtxt(
    'data/fig2a.txt', unpack=True, delimiter=',')
U1, Emax  = np.loadtxt(
    'data/fig2b.txt', unpack=True, delimiter=',')
U2, gtmax = np.loadtxt(
    'data/fig2c.txt', unpack=True, delimiter=',')

# Fig. 3
Tasolid, E1solid, E2solid, E3solid = np.loadtxt(
    'data/fig3a-eb.txt' , unpack=True, delimiter=',')
Tadashed, E1dashed, E2dashed, E3dashed = np.loadtxt(
    'data/fig3a-semi.txt', unpack=True, delimiter=',', dtype=np.complex_)
Tbsolid, E1bsolid, E2bsolid, E3bsolid, E4bsolid = np.loadtxt(
    'data/fig3b-eb.txt', unpack=True, delimiter=',')
Tbdashed, E1bdashed, E2bdashed, E3bdashed = np.loadtxt(
    'data/fig3b-semi.txt', unpack=True, delimiter=',', dtype=np.complex_)
Tc, E1c, E2c, E3c, E4c = np.loadtxt(
    'data/fig3c.txt', unpack=True, delimiter=',')

# Fig. 4
Tgt, EB1, EB2, EB3, EB4 = np.loadtxt(
    'data/fig4a.txt', unpack=True, delimiter=',')
Tgtergo, Ergo1, Ergo2, Ergo3, Ergo4 = np.loadtxt(
    'data/fig4a-ergo.txt', unpack=True, delimiter=',')
TgtPB, PB1, PB2, PB3, PB4 = np.loadtxt(
    'data/fig4d.txt', unpack=True, delimiter=',')
Tgta, EA1, EA2, EA3,EQ= np.loadtxt(
    'data/fig4c.txt', unpack=True, delimiter=',')
Tgtrat, R1, R2, R3,RQ= np.loadtxt(
    'data/fig4b.txt', unpack=True, delimiter=',')

# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 18
plt.rcParams['axes.linewidth'] = 2
```

```python
# POLISHED FIGURE 2

# Create figure object and store it in a variable called 'fig'
fig = plt.figure(figsize=(4, 7))

# Add two axes objects to create a paneled figure
ax2a = fig.add_axes([0.00, 0.65, 1.00, 0.35])
ax2b = fig.add_axes([0.00, 0.30, 1.00, 0.25])
ax2c = fig.add_axes([0.00, 0.00, 1.00, 0.25])


# Edit the major and minor ticks of the x and y axes
ax2a.xaxis.set_tick_params(which='major', size=8, width=1.5, direction='in', top='on')
ax2a.xaxis.set_tick_params(which='minor', size=5, width=1.5, direction='in', top='on')
ax2a.yaxis.set_tick_params(which='major', size=8, width=1.5, direction='in', right='on')
ax2a.yaxis.set_tick_params(which='minor', size=5, width=1.5, direction='in', right='on')
ax2b.xaxis.set_tick_params(which='major', size=8, width=1.5, direction='in', top='on')
ax2b.xaxis.set_tick_params(which='minor', size=5, width=1.5, direction='in', top='on')
ax2b.yaxis.set_tick_params(which='major', size=8, width=1.5, direction='in', right='on')
ax2b.yaxis.set_tick_params(which='minor', size=5, width=1.5, direction='in', right='on')
ax2c.xaxis.set_tick_params(which='major', size=8, width=1.5, direction='in', top='on')
ax2c.xaxis.set_tick_params(which='minor', size=5, width=1.5, direction='in', top='on')
ax2c.yaxis.set_tick_params(which='major', size=8, width=1.5, direction='in', right='on')
ax2c.yaxis.set_tick_params(which='minor', size=5, width=1.5, direction='in', right='on')

# Edit the major and minor tick locations of x and y axes
ax2a.xaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
ax2a.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
ax2a.yaxis.set_major_locator(mpl.ticker.MultipleLocator(2))
ax2a.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(1))
ax2b.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.01))
ax2b.yaxis.set_major_locator(mpl.ticker.MultipleLocator(2))
ax2b.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(1))
ax2c.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.01))
ax2c.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))

# Add the x and y-axis labels
ax2a.set_xlabel(r'$T$ ($gt/\pi$)'           , labelpad=1.5)
ax2a.set_ylabel(r'$E_B/\omega_0$'           , labelpad=1.0)
ax2b.set_ylabel(r'$E_\mathrm{max}/\omega_0$', labelpad=1.0)
ax2c.set_xlabel(r'$U/\omega_0$'             , labelpad=1.5)
ax2c.set_ylabel(r'$gt_\mathrm{max}/\pi$'    , labelpad=12.5)

# Set the axis limits
ax2a.set_xlim(0    , 6 )
ax2a.set_ylim(0    , 12)
ax2b.set_xlim(-0.01, 0.16)
ax2b.set_ylim(0.5  , 11.5)
ax2c.set_xlim(-0.01, 0.16)
ax2c.set_ylim(0    , 6)

# Plot Figure 2a
ax2a.text(-1.1, 11.5, '(a)')
ax2a.plot(time, ener0, linewidth=2, color='C0')
ax2a.text(4.1, 10  , r'$U/\omega_0 = 0$'    , color='C0', fontsize=12)
ax2a.plot(time, ener1, linewidth=2, color='C1')
ax2a.text(4.1, 6.50, r'$U/\omega_0 = 0.005$', color='C1', fontsize=12)
ax2a.plot(time, ener2, linewidth=2, color='C2')
ax2a.text(4.1, 4.00, r'$U/\omega_0 = 0.02$ ', color='C2', fontsize=12)
ax2a.plot(time, ener3, linewidth=2, color='C3')
ax2a.text(4.1, 2.51, r'$U/\omega_0 = 0.05$ ', color='C3', fontsize=12)
ax2a.plot(time, ener4, linewidth=2, color='C4')
ax2a.text(4.1, 1.43, r'$U/\omega_0 = 0.15$ ', color='C4', fontsize=12)
ax2a.plot(time, ener5, linewidth=2, color='C5')
ax2a.text(4.1, 0.60, 'Qubit'                , color='C5', fontsize=12)

# Plot Figure 2b
ax2b.text(-0.04, 11, '(b)')
ax2b.plot(U1, Emax, linestyle='None', marker="*", markersize=14, color='black')
ax2b.grid(True)

# Plot Figure 2c
ax2c.text(-0.04, 5.7, '(c)')
ax2c.plot(U2, gtmax, linestyle='None', marker="*", markersize=14, color='black')
ax2c.grid(True)

# Save figure
plt.savefig('Fig2.pdf',bbox_inches='tight')
plt.show()
```
